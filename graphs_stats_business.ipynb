```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#data import and dropping duplicates
def importData(filename):
    """
    Function to import data & dropping duplicates
    :param filename: <string> file input name (format .csv)
    :return df: <pandas dataframe> sampel data
    """

    # read data
    df = pd.read_csv(filename, encoding='utf-8')
    print("Data asli            : ", df.shape, "- (#observasi, #kolom)")

    # drop duplicates
    df = df.drop_duplicates()
    print("Data setelah di-drop : ", df.shape, "- (#observasi, #kolom)")

    return df

```

```python
data = 'CO2 Emissions_Canada.csv'
```

```python
emission_data = importData(filename = data)
```

```python
#changing the fuel type description to be more meaningful

emission_data["Fuel Type"] = np.where(emission_data["Fuel Type"]=="Z", "Premium Gasoline", emission_data["Fuel Type"])
emission_data["Fuel Type"] = np.where(emission_data["Fuel Type"]=="X", "Regular Gasoline", emission_data["Fuel Type"])
emission_data["Fuel Type"] = np.where(emission_data["Fuel Type"]=="D", "Diesel", emission_data["Fuel Type"])
emission_data["Fuel Type"] = np.where(emission_data["Fuel Type"]=="E", "Ethanol(E85)", emission_data["Fuel Type"])
emission_data["Fuel Type"] = np.where(emission_data["Fuel Type"]=="N", "Natural Gas", emission_data["Fuel Type"])
```

```python
#Reducing Transmission types

emission_data["Transmission"] = np.where(emission_data["Transmission"].isin(["A4", "A5", "A6", 
                                                                "A7", "A8", "A9", "A10"]), "Automatic", emission_data["Transmission"])

emission_data["Transmission"] = np.where(emission_data["Transmission"].isin(["AM5", "AM6", 
                                                                "AM7", "AM8", "AM9"]), "Automated Manual", emission_data["Transmission"])

emission_data["Transmission"] = np.where(emission_data["Transmission"].isin(["AS4", "AS5", 
                                                                  "AS6", "AS7", "AS8", "AS9", "AS10"]), "Automatic with Select Shift", 
                                                                  emission_data["Transmission"])

emission_data["Transmission"] = np.where(emission_data["Transmission"].isin(["AV", "AV6",
                                                                  "AV7", "AV8", "AV10"]), "Continuously Variable", 
                                                                  emission_data["Transmission"])

emission_data["Transmission"] = np.where(emission_data["Transmission"].isin(["M5", "M6",
                                                                  "M7"]), "Manual", emission_data["Transmission"])
```

```python
#drop unused columns

drop_columns = ["Make", "Model", "Fuel Consumption City (L/100 km)", 
                "Fuel Consumption Hwy (L/100 km)", "Fuel Consumption Comb (mpg)"]
                
emission = emission_data.drop(drop_columns, axis=1)
```

```python
#Checking Fuel Type sample distribution

plt.figure(figsize=(10, 5))
fuel_sort = data['Fuel Type'].sort_values().unique()
fuel_freq = data['Fuel Type'].value_counts().sort_index(axis=0)

df = pd.DataFrame({'Category':fuel_sort,
                  'Frekuensi':fuel_freq})

df = df.sort_values(by='Frekuensi', ascending=False)

plt.bar(df['Category'], df['Frekuensi'], color=['steelblue', 'orange', 'green', 'red'])
plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Checking Transmission sample distribution

plt.figure(figsize=(10, 5))
trans_sort = data['Transmission'].sort_values().unique()
trans_freq = data['Transmission'].value_counts().sort_index(axis=0)

df = pd.DataFrame({'Category':trans_sort,
                  'Frekuensi':trans_freq})

df = df.sort_values(by='Frekuensi', ascending=False)

plt.bar(df['Category'], df['Frekuensi'], color=['steelblue', 'orange', 'green', 'red', 'purple'])
plt.xlabel('Transmission', fontsize=15)

plt.ylabel('Frequency',  fontsize=15)
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.show()
```

```python
#Checking vehicle class sample distribution

plt.figure(figsize=(20, 5))
vehicle_sort = data['Vehicle Class'].sort_values().unique()
vehicle_freq = data['Vehicle Class'].value_counts().sort_index(axis=0)

df = pd.DataFrame({'Category':vehicle_sort,
                  'Frekuensi':vehicle_freq})

df = df.sort_values(by='Frekuensi', ascending=False)

plt.bar(df['Category'], df['Frekuensi'])
plt.xlabel('Vehicle Class', fontsize=15)
plt.xticks(rotation=45)
plt.ylabel('Frequency',  fontsize=15)
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)
plt.show()
```
